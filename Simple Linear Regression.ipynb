```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
plt.figure(figsize=(6,2))
plt.boxplot(data1["daily"], vert = False)
```

```python
sns.histplot(data1["daily"],kde=True,color='orange',bins=30)

#show the plot
plt.show()
```

```python
fig,axes = plt.subplots(2,1,figsize=(8,6), gridspec_kw={'height_ratios':[1,3]})
#plot  the boxplot in the first (top) subplot
sns.boxplot(data=data1["daily"],ax= axes[0],color='skyblue',width=0.5,orient = 'h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Ozone Levels")


#plot the histogram with KDE curve in the second (bottom) subplot
sns.histplot(data1["daily"],kde=True,ax=axes[1],color='purple',bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Ozone Levels")
axes[1].set_ylabel("Frequency")

# aadjust layout for better spacing
plt.tight_layout()

#show the plot
plt.show()
```

## observations
- a high correlation coefficient is observed between daily and sunday 

```python
import seaborn as sns
sns.histplot(data1['sunday'], kde = True,stat='density',)
plt.show()
```

```python
# Build regression model

import statsmodels.formula.api as smf
model = smf.ols("sunday~daily",data = data1).fit()
```

```python
model.summary()
```

```python
x = data1["daily"].values
y = data1["sunday"].values
plt.scatter(x, y, color = "m", marker = "o", s = 30)
b0 = 13.84
b1 =1.33
# predicted response vector
y_hat = b0 + b1*x
 
# plotting the regression line
plt.plot(x, y_hat, color = "g")
  
# putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Plot the linear regression line using seaborn regplot() method
sns.regplot(x="daily", y="sunday", data=data1)
plt.xlim([0,1250])
plt.show()
```
