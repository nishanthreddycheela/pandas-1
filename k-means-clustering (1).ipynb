#  Clustering - Divide the Universities into groups (Clusters)

```python
Univ = pd.read_csv("Universities.csv")
Univ
```

```python
# Boxplot for daily column

plt.figure(figsize=(6,3))
plt.title("Box plot for sat")
plt.boxplot(Univ["SAT"],vert = False)
plt.show()
```

```python
sns.histplot(Univ['SAT'],kde = True, stat='density')
plt.show()
```

```python
Univ.describe()
```

```python
# read all numeric columns into Univ1
Univ1 = Univ.iloc[:,1:]
```

```python
Univ1
```

```python
cols = Univ1.columns
cols
```

```python
#  standardisation function

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Univ_df = pd.DataFrame(scaler.fit_transform(Univ1),columns = cols)
scaled_Univ_df
# scaler.fit_transform(Univ1)
```

```python
# build 3 clusters using kmeans cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=0) # specify 3 clusters
clusters_new.fit(scaled_Univ_df)
```

```python
# print the cluster labels
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
# Assign clusters to the Univ data set
Univ['clusterid_new'] = clusters_new.labels_
```

```python
Univ
```

```python

Univ.sort_values(by = "clusterid_new")
```

```python
# Use groupby() to find aggregated (mean) values in each other
Univ.iloc[:,1:].groupby("clusterid_new").mean()
```

# observations
- cluster 2 appears to be the top rated univesities cluster as the cut off score, top10,sfrratio paramenter mean values are highest
- cluster 1 appears to occupy the middle level rated universities
- cluseter 0 comes as the lower level rated universities

```python
Univ[Univ['clusterid_new']==0]
```

```python
# Finding optimal k value using elbow plot
wcss = []
for i in range(1, 20):

    Kmeans = KMeans(n_clusters=i,random_state=0)
    Kmeans.fit(scaled_Univ_df)
    #Kmeans.fit(Univ1)
    wcss.append(Kmeans.inertia_)

print(wcss)
plt.plot(range(1,20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

# observations
- from the above graph we can choose k = 3 or 4 which indicates the elbow joint that is the rate of change of slope decreases
